## Problem 2: Visualizing CIFAR-10

The CIFAR-10 dataset contains 60,000 RGB images from 10 categories. It can be found here:
https://www.cs.toronto.edu/~kriz/cifar.html. It can alternatively be loaded via HuggingFace
datasets (make sure to `pip install datasets` before importing).<br>
<br>Using the first CIFAR-10 training batch file, display the first three images from each of the 10
categories as a 3 × 10 image array. The images are stored as rows, and you will need to
reshape them into 32 × 32 × 3 images if you load up the raw data yourself. It is okay to use the
PyTorch toolbox for loading them or you can roll your own.


### HuggingFace

```python
dataset = load_dataset('cifar10')
```

```python
def huggingface_cifar10():
    """
    
    """
    df = pd.DataFrame(dataset["train"])
    img_by_cat = []
    for i in range(10):
        img_by_cat.append(list(df.loc[df['label']==i][:3]["img"]))

    fig, axs = plt.subplots(3, 10, figsize=(15, 6))
    # Plot the images from the dictionary
    for i in range(10):
        for j in range(3):
            ax = axs[j][i]
            ax.imshow(img_by_cat[i][j])
            ax.axis('off')

    return plt.show()
```

```python
huggingface_cifar10()
```

### CIFAR-10


```python
# Load the CIFAR-10 dataset
cifar10 = datasets.CIFAR10(root='path/to/data', train=True, download=True, transform=transforms.ToTensor())


def pytorch_cifar():
    """Pytorch_cifar10
    Loads in the cifar10 dataset from the Pytorch 


    """

    # Create a subplot of 3 rows and 10 columns
    fig, axs = plt.subplots(3, 10, figsize=(15, 6))

    # Create a dictionary to store the first three images from each category
    categories = {i: [] for i in range(10)}

    # Iterate through the dataset
    for data, label in cifar10:
        if len(categories[label]) < 3:
            categories[label].append(data)

    # Plot the images from the dictionary
    for i, category in enumerate(categories):
        for j, img in enumerate(categories[category]):
            ax = axs[j][i]
            ax.imshow(img.permute(1, 2, 0))
            ax.axis('off')
            ax.set_title(cifar10.classes[category])

    return plt.show()
```

```python
pytorch_cifar()
```

## Problem 2: Gaussian Function 

Write a function called gaussian(n, m) that returns an n × m NumPy array, each entry of which is
a random number drawn from the standard normal distribution (Gaussian with mean 0 variance
1). Generate a 10x10 grid of 2-D points using this function and then make a scatter plot of the
points using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

#Creates the array 
def gaussian(n: int, m:int):
    """Plot gaussian function 

    Args:
    n: initial integer value for the array 
    m: initial intger value for the array

    Returns:
    Scatter plot of the n by m by 2 array 
    """
    arr=  np.random.randn(n, m, 2)

    #Formats the data 
    x, y = arr[:, :, 0].flatten(), arr[:, :, 1].flatten()

    #PLots the points 
    plt.scatter(x, y)

    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("100 Gaussian Points")
    #Returns the final result 
    return plt.show()


#Calls the data
gaussian(10,10)
```
